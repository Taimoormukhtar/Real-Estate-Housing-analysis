## **Data Cleaning**

Data cleaning and preprocessing is the major step in Data analysis. It includes steps from handling missing data to removing outliers and detecting duplicates

##### **Naming the Columns Correctly**

```python
#Ensure Proper Column Names
df_housing.rename(columns={'Garage Yr Blt': 'Garage Year Built','Yr Sold':'Year Sold','TotRmsAbvGrd':'Total Rooms Above Grade','Exter Qual':'Exterior Quality',
                   'Exter Cond':'Exterior Condition','Lot Config':'Lot Configuration','Overall Qual':'Overall Quality','Overall Cond':'Overall Condition',
                    'Roof Matl':'Roof material','Mas Vnr Type':'Masonry veneer type','Mas Vnr Area':'Masonry veneer area SF','Bsmt Qual':'Basement Quality',
                    'Bsmt Cond':'Basement Condition','Bsmt Exposure':'Basement Exposure','BsmtFin Type 1':'Rating of basement finished area',
                    'BsmtFin SF 1':'Type 1 Basement finished square feet','BsmtFin Type 2':'Rating of basement finished area(multiple)',
                    'BsmtFin SF 2':'Type 2 Basement finished square feet','Bsmt Unf SF':'Unfinished SF of basement area',
                    'Total Bsmt SF':'Total Basement SF','1st Flr SF':'1st Floor SF','2nd Flr SF':'2nd Floor SF','Gr Liv Area':'Ground Living Area',
                    'Low Qual Fin SF':'Low Quality Finished SF','Bsmt Full Bath':'Basement Full Bath','Bsmt Half Bath':'Basement Half Bath',
                    'Bedroom AbvGr':'Bedroom Above Ground','Kitchen AbvGr':'Kitchen Above Ground','Kitchen Qual':'Kitchen Quality',
                    'TotRms AbvGrd':'Total Rooms Above Ground','Fireplace Qu':'Fireplace Quality',
                    'Garage Qual':'Garage Quality','3-Ssn Porch':'Three season porch area in SF','Pool QC':'Pool Quality',
                    'Misc Feature':'Miscellaneous feature','Misc Val':'$ Value of Miscellaneous feature',
                    'Mo Sold':'Month Sold','Year Remod/Add':'Year remodel/addition'}, inplace=True)
```

##### **DataTypes Correction**

```python
#Checking Datatypes for All Columns
column_data_types = df_housing.dtypes
for column, data_type in column_data_types.items():
    print(f"Column '{column}': {data_type}")
```

There are many wrong datatypes present. Notice the below columns are of object datatype. We need to convert these columns to Datetime datatype to ensure correct analysis.

*   Year built
*   Year remodal/addition
*   Garage Year Built
*   Year Sold

Following of the columns are present as integers. They need to be converted to object datatype.

*  Order_1
*  Order_2
*  PID
*  MS Subclass_1
*  MS Subclass_2

```python
#Converting Datatypes
df_housing['Year Built'] = pd.to_datetime(df_housing['Year Built'],format='%Y')
df_housing['Year remodel/addition'] = pd.to_datetime(df_housing['Year remodel/addition'],format='%Y')
df_housing['Garage Year Built'] = pd.to_datetime(df_housing['Garage Year Built'],format='%Y')
df_housing['Year Sold'] = pd.to_datetime(df_housing['Year Sold'],format='%Y')

df_housing['Order_1'] = df_housing['Order_1'].astype('object')
df_housing['Order_2'] = df_housing['Order_2'].astype('object')
df_housing['PID'] = df_housing['PID'].astype('object')
df_housing['MS SubClass_1'] = df_housing['MS SubClass_1'].astype('object')
df_housing['MS SubClass_2'] = df_housing['MS SubClass_2'].astype('object')
```

```python
#Checking Again Datatypes to confirm
selected_columns=['Order_1','Order_2', 'PID', 'MS SubClass_2','MS SubClass_1','Year Built', 'Year remodel/addition',
                  'Garage Year Built','Year Sold']
df_housing[selected_columns].dtypes
```

##### **Syntax Correction**

Month Sold column have integer values and it needs to be converted to month names instead.



```python
#Converting integer values to Month Name
month_mapping = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'}
df_housing['Month Sold'] = df_housing['Month Sold'].map(month_mapping)
```

##### **Identify Outliers**

```python
#Removing Outliers
from scipy.stats import zscore
numerical_columns = df_housing.select_dtypes(include=[np.number]).columns
z_scores = zscore(df_housing[numerical_columns])
threshold = 3
outliers = (np.abs(z_scores) > threshold).any(axis=1)
clean_house_data = df_housing[~outliers]
clean_house_data.shape
```

The logic behind using Z-score as a threshold value for removing outliers is rooted in the idea that values with high Z-scores (far from the mean) may be considered unusual or anomalous and can be treated as outliers. Using
±3 as a threshold aligns with the idea that extreme values beyond a few standard deviations may be considered outliers.

## **Exploratory Data Analysis**

After performing all the necessary steps. The last step is Exploratory data analysis

##### **Value Counts for Categorical Columns**

```python
selected_columns = ['MS SubClass_1','Basement Quality','Exterior Condition','Month Sold','Lot Configuration','House Style',
                    'Pool Quality','Central Air','Garage Type']
# Calculate the number of rows needed
num_rows = (len(selected_columns) + 2) // 3
# Create subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, num_rows * 5))
# Flatten the axes for easier iteration
axes = axes.flatten()
# Plot countplots for each selected column
for i, column in enumerate(selected_columns):
    counts = df_housing[column].value_counts().sort_values(ascending=False)
    sns.countplot(x=column, data=df_housing, order=counts.index, ax=axes[i])
    axes[i].set_title(f'Countplot for {column}')
    axes[i].tick_params(axis='x', rotation=90)
# Adjust layout
plt.tight_layout()
plt.show()
```

Based on these countplots, one can figure out certain features that are reccuring in the dataset.

##### **Time Series Analysis**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(df_housing,x='Year Built',y='Overall Quality',errorbar=None,color='blue',lw=1.5,label='Overall Quality')
sns.lineplot(df_housing,x='Year Built',y='Overall Condition',errorbar=None,lw=1.5,label='Overall Condition'
             ,color='black',ls='--')
plt.ylabel('Overall Quality and Condition')
plt.title('Overall Quality and Condition by Year Built',fontsize=12)
plt.show()
```

This dual line chart tells that the Overall Quality is increasing as we got into 21st Century while the opposite is happening with Overall Conditions.

```python
# Adding Rolling Mean SalePrice and Rolling Standard Deviation of SalePrice as new columns
df_housing['Mean_SalePrice'] = df_housing['SalePrice'].rolling(3).mean()
df_housing['Min_SalePrice'] = df_housing['SalePrice'].rolling(3).min()
df_housing['Max_SalePrice'] = df_housing['SalePrice'].rolling(3).max()
```

A rolling mean is simply the mean of a certain number of previous periods in a time series. The 3 written in round brackets is the window size. You can edit the window size based on the type of results you want. For example, if one want to observe long term fluctuations, larger window size is preferred.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(df_housing,x='Year Sold',y='Max_SalePrice',errorbar=None,color='red',ls='-.',lw=1.5,label='Max SalePrice')
sns.lineplot(df_housing,x='Year Sold',y='Mean_SalePrice',errorbar=None,lw=1.5,label='Mean SalePrice',color='black')
sns.lineplot(df_housing,x='Year Sold',y='Min_SalePrice',errorbar=None,lw=1.5,label='Min SalePrice',color='blue',ls='--')
plt.ylabel('Sale Price')
plt.title('Observing trends in SalePrice',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(df_housing,x='Year Built',y='Garage Area',errorbar=None,color='red',ls='-.',lw=1.5,label='Garage Area')
sns.lineplot(df_housing,x='Year Built',y='Ground Living Area',errorbar=None,lw=1.5,label='Ground Living Area',color='black')
plt.ylabel('Area in Square Feet')
plt.title('Garage Area and Ground Living Area Over the Time',fontsize=12)
plt.show()
```

##### **Correlation Analysis**

```python
plt.figure(figsize=(10,6))
selected_columns = df_housing[['Overall Condition','Overall Quality',"Lot Area",'Garage Area','SalePrice','Ground Living Area',
                   'Total Rooms Above Ground']]
df_corr=selected_columns.corr()
sns.heatmap(df_corr,cmap='crest',annot=True,linecolor='black',linewidth=0.1,fmt='0.2g',square=True)
plt.show()
```

This heatmap tells correlation between different variables. The values are present between -1 to 1. -1 tells strong negative correlation and 1 gives strong positive correlation. Any value near to zero means no correlation. For example, there is strong positive correlation between Sale Price and Overall Quality.

##### **Relational Plots**

```python
sns.scatterplot(df_housing,x='Ground Living Area',y='SalePrice',hue='Overall Condition',
                size='Overall Condition',palette='viridis')
plt.title('Does Living Area Affects SalePrice?',fontsize=12)
plt.grid()
plt.show()
```

```python
sns.scatterplot(df_housing,x='Garage Area',y='SalePrice',hue='Overall Condition',size='Overall Condition',palette='viridis')
plt.title('Does Garage Area Affects SalePrice?',fontsize=12)
plt.grid()
plt.show()
```

By observing the scatterplots, it is the Ground Living Area that affects the Sale Price more rather than Garage Area. However, there is small linearity difference between these two variables for SalePrice becuase Living Area and Garage Area are correlated. Lets observe that too to understand better

```python
sns.regplot(df_housing,x='Garage Area',y='Ground Living Area',scatter_kws={'s': 10, 'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Does Garage Area and Ground Living Area are positively correlated?',fontsize=11)
plt.grid()
plt.show()
```

Seems like Garage Area and Ground Living Area are correlated. There correlation value is 0.5
